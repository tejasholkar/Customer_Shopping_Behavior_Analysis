```python
df['Review Rating'] = df.groupby('Category') ['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a columns age_group
labels = ['Young Adult' , 'Adult' , 'Middle-age' , 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4 , labels=labels)
```

```python
df[['age' , 'age_group']].head(10)
```

```python
#create a column purchase_frequency_days 
frequency_mapping = {
    'Fortnightly' : 14 ,
    'Weekly' : 7 ,
    'Monthly' : 30 ,
    'Quarterly' : 90 ,
    'Bi-Weekly' : 14 ,
    'Annually' : 365 ,
    'Every 3 months' : 90,
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied' , 'promo_code_used']].head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("customer_shopping_behavior.csv")

print(df.head())
print(df.columns)
```

```python
plt.figure(figsize=(10,5))

category_sales = df.groupby("Category")["Purchase Amount (USD)"].sum().sort_values(ascending=False)

category_sales.plot(kind="bar", color="skyblue")

plt.title("Total Purchase Amount by Category")
plt.xlabel("Category")
plt.ylabel("Total Purchase Amount (USD)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Purchase Amount (USD)"], bins=20, edgecolor="black")

plt.title("Distribution of Purchase Amount")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Number of Customers")
plt.grid(alpha=0.3)

plt.show()
```

```python
top_items = df["Item Purchased"].value_counts().head(10)

plt.figure(figsize=(10,6))

top_items.plot(kind="barh", color="orange")

plt.title("Top 10 Most Purchased Items")
plt.xlabel("Number of Purchases")
plt.ylabel("Item Purchased")
plt.gca().invert_yaxis()

plt.show()
```

```python
plt.figure(figsize=(7,5))

numeric_df = df.select_dtypes(include=["int64","float64"])

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["Review Rating"],
    df["Purchase Amount (USD)"],
    alpha=0.6
)

plt.title("Purchase Amount vs Review Rating")
plt.xlabel("Review Rating")
plt.ylabel("Purchase Amount (USD)")
plt.grid(alpha=0.3)

plt.show()
```
